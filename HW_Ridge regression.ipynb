```python
import numpy as np
import pandas as pd
#from sklearn import linear model
from sklearn.model_selection import train_test_split,KFold, cross_val_score, cross_val_predict
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import  LinearRegression, Ridge, Lasso, RidgeCV, LassoCV
```

```python
df=pd.read_csv("houses_train.csv")
df
```

```python
df['Unnamed: 0'].nunique()
```

```python
df['region'].unique()
```

```python
df['street'].value_counts()
```

```python
y=df['price']
y[:5]
```

```python
X=df.drop(['Unnamed: 0','price','street','region','url'], axis=1)
X.head()
```

```python
X['district'].value_counts()
```

```python
X=pd.get_dummies(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
lm = LinearRegression()
lm.fit(X_train, y_train)
lm.coef_, lm.intercept_
```

```python
preds=lm.predict(X_test)
preds
```

```python
print(np.sqrt(mean_squared_error(y_train, lm.predict(X_train))))
print(np.sqrt(mean_squared_error(y_test, lm.predict(X_test))))
```

## Ridge regression 

```python
rr=Ridge(alpha=13)
rr.fit(X_train,y_train)
rr.coef_, rr.intercept_
```

```python
rr.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(y_train, rr.predict(X_train))))
print(np.sqrt(mean_squared_error(y_test, rr.predict(X_test))))
```

## Lasso Regression

```python
lr=Lasso(alpha=13)
lr.fit(X_train,y_train)
lr.coef_, lr.intercept_
```

```python
rr.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(y_train, lr.predict(X_train))))
print(np.sqrt(mean_squared_error(y_test, lr.predict(X_test))))
```

## Cross-validation + Ridge Regression with automatically calculated alpha

```python
ridge = RidgeCV(normalize=True, scoring = 'neg_mean_squared_error', cv=10)
```

```python
scores = cross_val_score(ridge, X, y,scoring = 'neg_mean_squared_error', cv=10)
np.mean(np.sqrt(np.abs(scores)))
```

```python
predictions = cross_val_predict(ridge, X, y, cv=10)
predictions
```

### Cross-validation + Ridge Regression without cross_val_predict

```python
kf = KFold(n_splits=10)
kf.get_n_splits(X)
```

```python
RMSE_test=[]
RMSE_train=[]
for tr_index, ts_index in kf.split(X):
    X_tr, X_ts = X.iloc[tr_index], X.iloc[ts_index]
    y_tr, y_ts = y[tr_index], y[ts_index]
    rr_cv=Ridge(0.01)
    rr_cv.fit(X_tr,y_tr)
    preds_tr=rr_cv.predict(X_tr)
    preds_ts=rr_cv.predict(X_ts)
    RMSE_test.append(float(np.sqrt(mean_squared_error(preds_ts, y_ts))))
    RMSE_train.append(float(np.sqrt(mean_squared_error(preds_tr, y_tr))))
#print ('Test RMSE:', RMSE_test)
#print ('Train RMSE:', RMSE_train)
print("Mean train RMSE(Ridge regression):", sum(RMSE_train)/len(RMSE_train))
print("Mean test RMSE(Ridge regression):", sum(RMSE_test)/len(RMSE_test))
```

### Cross-validation + Ridge

```python
r=Ridge(alpha=0.01)
scores_rg=cross_val_score(r,X,y, scoring = 'neg_mean_squared_error',cv=10)
print('Lasso scores:', scores_rg)
y_rg=cross_val_predict(r, X, y, cv=10)
print('Predictions:', y_rg)
RMSE_mean_Ridge=np.mean(np.sqrt(np.abs(scores_rg)))
print("Mean RMSE(Ridge regression):", RMSE_mean_Ridge)
```

### Cross-validation + Lasso

```python
lasso=Lasso(alpha=0.3, max_iter=100000)
scores_ls=cross_val_score(lasso,X,y, scoring = 'neg_mean_squared_error',cv=10)
print('Lasso scores:', scores_ls)
y_ls=cross_val_predict(lasso, X, y, cv=10)
print('Predictions:', y_ls)
```

```python
RMSE_mean_Lasso=np.mean(np.sqrt(np.abs(scores_ls)))
print("Mean RMSE(Lasso regression):", RMSE_mean_Lasso)
```
