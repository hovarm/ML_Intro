```python
import numpy as np
import pandas as pd
df=pd.read_csv('telecom.csv')
```

```python
df.nunique()
```

```python
df.head()
```

```python

df['Churn'].replace({'Yes':1,'No':0}, inplace=True)
```

```python
df1=pd.get_dummies(df[['gender','Partner','Dependents','PhoneService','MultipleLines','PaperlessBilling','InternetService','OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingMovies','StreamingTV','Contract','PaymentMethod']])
```

```python
df1.head()
```

```python
y=df["Churn"]
```

```python
df=df.drop(['Churn','customerID','gender','Partner','Dependents','PhoneService','MultipleLines','PaperlessBilling','InternetService','OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingMovies','StreamingTV','Contract','PaymentMethod'], axis=1)
```

```python
df['TotalCharges'] = df['TotalCharges'].apply(pd.to_numeric,errors='coerce')
```

```python
df=pd.merge(df, df1, left_index=True, right_index=True)
df.head()
```

```python
is_NaN=df.isnull()
row_has_NaN=is_NaN.any(axis=1)
df[row_has_NaN]
```

```python
df.fillna(value=np.mean(df['TotalCharges']), inplace=True)
```

```python
df.head()
```

# Build 5 classification models

logistic regression, decision tree, random forest, knn, svm

#### tips
<br>
1. use cross validation
<br>
2. use same splits for all classifiers (to have correct comparison)
<br>
3. build a table to show statistics for all classifiers (at least - accuracy, precision and recall. Any more component would be a +. For train - at least accuracy)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
```

### Logistic regression

```python
kf = KFold(n_splits=5)
lr_tr_accuracy = []
lr_tr_precision = []
lr_tr_recall=[]
lr_ts_accuracy = []
lr_ts_precision = []
lr_ts_recall=[]

for train_index, test_index in kf.split(data):
    X_train, X_test = data.iloc[train_index], data.iloc[test_index]
    y_train, y_test = churn[train_index], churn[test_index]
    
    lr = LogisticRegression(max_iter=500).fit(X_train, y_train)
    predictions = lr.predict(X_test)
    predictions_train = lr.predict(X_train)
    
    test_accuracy=accuracy_score(y_test,predictions)
    test_precision=precision_score(y_test,predictions)
    test_recall=recall_score(y_test,predictions)
    
    #print("TEST ACCURACY:", test_accuracy, "TEST PRECISION", test_precision, "TEST RECALL", test_recall)
    
    train_accuracy=accuracy_score(y_train,predictions_train)
    train_precision=precision_score(y_train,predictions_train)
    train_recall=recall_score(y_train,predictions_train)
    
    #print("TRAIN ACCURACY:", train_accuracy, "TRAIN PRECISION", train_precision, "TRAIN RECALL", train_recall)
    
    lr_tr_accuracy.append(train_accuracy)
    lr_tr_precision.append(train_precision)
    lr_tr_recall.append(train_recall)
    
    lr_ts_accuracy.append(test_accuracy)
    lr_ts_precision.append(test_precision)
    lr_ts_recall.append(test_recall)

#print('\n')
print("MEAN TEST ACCURACY:", np.mean(lr_ts_accuracy), "MEAN TEST PRECISION", np.mean(lr_ts_precision), "MEAN TEST RECALL", np.mean(lr_ts_recall))
print("MEAN TRAIN ACCURACY:", np.mean(lr_tr_accuracy), "TRAIN PRECISION", np.mean(lr_tr_precision), "TRAIN RECALL", np.mean(lr_tr_recall))
```

### Decision Tree

```python
kf = KFold(n_splits=5)
dt_tr_accuracy = []
dt_tr_precision = []
dt_tr_recall=[]
dt_ts_accuracy = []
dt_ts_precision = []
dt_ts_recall=[]

for train_index, test_index in kf.split(data):
    X_train, X_test = data.iloc[train_index], data.iloc[test_index]
    y_train, y_test = churn[train_index], churn[test_index]
    
    dt = DecisionTreeClassifier().fit(X_train, y_train)
    predictions = dt.predict(X_test)
    dt_predictions_train = dt.predict(X_train)
    
    test_accuracy=accuracy_score(y_test,predictions)
    test_precision=precision_score(y_test,predictions)
    test_recall=recall_score(y_test,predictions)
    
    #print("TEST ACCURACY:", test_accuracy, "TEST PRECISION", test_precision, "TEST RECALL", test_recall)
    
    train_accuracy=accuracy_score(y_train,predictions_train)
    train_precision=precision_score(y_train,predictions_train)
    train_recall=recall_score(y_train,predictions_train)
    
    #print("TRAIN ACCURACY:", train_accuracy, "TRAIN PRECISION", train_precision, "TRAIN RECALL", train_recall)
    
    dt_tr_accuracy.append(train_accuracy)
    dt_tr_precision.append(train_precision)
    dt_tr_recall.append(train_recall)
    
    dt_ts_accuracy.append(test_accuracy)
    dt_ts_precision.append(test_precision)
    dt_ts_recall.append(test_recall)

#print('\n')
print("MEAN TEST ACCURACY:", np.mean(dt_ts_accuracy), "MEAN TEST PRECISION", np.mean(dt_ts_precision), "MEAN TEST RECALL", np.mean(dt_ts_recall))
print("MEAN TRAIN ACCURACY:", np.mean(dt_tr_accuracy), "TRAIN PRECISION", np.mean(dt_tr_precision), "TRAIN RECALL", np.mean(dt_tr_recall))
```

### Random Forest

```python
kf = KFold(n_splits=5)
rf_tr_accuracy = []
rf_tr_precision = []
rf_tr_recall=[]
rf_ts_accuracy = []
rf_ts_precision = []
rf_ts_recall=[]

for train_index, test_index in kf.split(data):
    X_train, X_test = data.iloc[train_index], data.iloc[test_index]
    y_train, y_test = churn[train_index], churn[test_index]
    
    rf = RandomForestClassifier(max_depth=5).fit(X_train, y_train)
    predictions = rf.predict(X_test)
    dt_predictions_train = rf.predict(X_train)
    
    test_accuracy=accuracy_score(y_test,predictions)
    test_precision=precision_score(y_test,predictions)
    test_recall=recall_score(y_test,predictions)
    
    #print("TEST ACCURACY:", test_accuracy, "TEST PRECISION", test_precision, "TEST RECALL", test_recall)
    
    train_accuracy=accuracy_score(y_train,predictions_train)
    train_precision=precision_score(y_train,predictions_train)
    train_recall=recall_score(y_train,dt_predictions_train)
    
    #print("TRAIN ACCURACY:", train_accuracy, "TRAIN PRECISION", train_precision, "TRAIN RECALL", train_recall)
    
    rf_tr_accuracy.append(train_accuracy)
    rf_tr_precision.append(train_precision)
    rf_tr_recall.append(train_recall)
    
    rf_ts_accuracy.append(test_accuracy)
    rf_ts_precision.append(test_precision)
    rf_ts_recall.append(test_recall)

#print('\n')
print("MEAN TEST ACCURACY:", np.mean(rf_ts_accuracy), "MEAN TEST PRECISION", np.mean(rf_ts_precision), "MEAN TEST RECALL", np.mean(rf_ts_recall))
print("MEAN TRAIN ACCURACY:", np.mean(rf_tr_accuracy), "TRAIN PRECISION", np.mean(rf_tr_precision), "TRAIN RECALL", np.mean(rf_tr_recall))
```

### Testing

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=1/6)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
lst=[LogisticRegression(max_iter=500),DecisionTreeClassifier(),RandomForestClassifier(max_depth=5),svm.SVC(kernel='linear'),KNeighborsClassifier(n_neighbors=5)]
for i in lst:
    model=i.fit(X_train, y_train)
    predictions = model.predict(X_test)
    predictions_train = model.predict(X_train)

    test_accuracy=accuracy_score(y_test,predictions)
    test_precision=precision_score(y_test,predictions)
    test_recall=recall_score(y_test,predictions)

    #print(str(i).split('(')[0], "TEST ACCURACY:", round(test_accuracy,3), "TEST PRECISION", round(test_precision,3), "TEST RECALL", round(test_recall,3))
    
    train_accuracy=accuracy_score(y_train,predictions_train)
    train_precision=precision_score(y_train,predictions_train)
    train_recall=recall_score(y_train,predictions_train)
          
    #print(str(i).split('(')[0], "TRAIN ACCURACY:", round(train_accuracy,3), "TRAIN PRECISION", round(train_precision,3), "TRAIN RECALL", round(train_recall,3))
    d = {str(i).split('(')[0]: [round(test_accuracy,3),round(train_accuracy,3),round(test_precision,3),round(train_precision,3),round(test_recall,3),round(train_recall,3)]}
    table=pd.DataFrame.from_dict(d, orient='index',columns=['Test Accuracy','Train Accuracy', 'Test Precision', 'Train Precision', 'Test Recall','Train Recall'])  
    table.append(d, ignore_index=True)
    print(table)
```
